```python
from bokeh.plotting import figure, output_file, show
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Specify the output HTML file
output_file('custom_plot.html')

# Create a new plot with a title and axis labels
p = figure(title='Custom Plot Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circle glyphs with customizations
p.circle(x, y, size=8, color='blue', alpha=0.5, legend_label='sin(x)')

# Customize plot appearance
p.title.text_color = 'darkgreen'
p.title.text_font_style = 'italic'
p.xaxis.axis_label_text_color = 'orange'
p.yaxis.axis_label_text_font_size = '14pt'

# Show the plot with legend
p.legend.title = 'Legend'
p.legend.location = 'top_left'

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource, Slider
import numpy as np

# Create data source and initial data
source = ColumnDataSource(data=dict(x=[], y=[]))

# Create a new plot with a title and specify height and width
plot = figure(title="Real-Time Plot", height=300, width=600)
plot.line(x='x', y='y', source=source)

# Create a slider widget
slider = Slider(start=0, end=10, value=0, step=0.1, title="Frequency")

# Define callback function to update plot
def update_data(attrname, old, new):
    freq = slider.value
    x = np.linspace(0, 10, 500)
    y = np.sin(freq * x)
    source.data = dict(x=x, y=y)

# Attach callback function to slider
slider.on_change('value', update_data)

# Create layout and add plot and slider
layout = plot

# Add layout to current document (curdoc)
curdoc().add_root(layout)
curdoc().title = "Real-Time Plot"
```

```python
# app.py

from flask import Flask, render_template
from bokeh.embed import components
from bokeh.plotting import figure
from bokeh.sampledata.iris import flowers
from bokeh.models import ColumnDataSource
import numpy as np

app = Flask(__name__)

@app.route('/')
def index():
    # Generate Bokeh plot
    p = figure(title="Iris Flower Scatter Plot", x_axis_label='Sepal Length', y_axis_label='Sepal Width')
    source = ColumnDataSource(flowers)
    p.circle('sepal_length', 'sepal_width', source=source, size=10, color='navy', alpha=0.5)

    # Embed plot into HTML
    script, div = components(p)
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
